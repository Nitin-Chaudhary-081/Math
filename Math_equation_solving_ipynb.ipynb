```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Gather points from the user
x_coords = []
y_coords = []

print("--- 4-Quadrant Custom Graph Maker ---")
try:
    num_points = int(input("How many points do you want to input? "))
except ValueError:
    print("Please enter a valid integer number.")
    exit()

for i in range(num_points):
    print(f"\n--- Point {i+1} ---")
    try:
        x =float(input(f"Enter X value for point {i+1}: "))
        y = float(input(f"Enter Y value for point {i+1}: "))
        x_coords.append(x)
        y_coords.append(y)
    except ValueError:
        print("Invalid input. Please enter numbers only.")
        exit()

# Clear any previous figures
plt.clf()

# 2. Draw a line connecting every dot in order and mark the dots
# 'marker=o' places dots, 'color=blue' draws the line, 'markerfacecolor=red' colors the dots red
plt.plot(x_coords, y_coords, color='blue', linewidth=2, marker='o',
         markerfacecolor='red', markersize=8, label='Connected Path')

# 3. Label each dot with its exact (X, Y) coordinate numbers
for i, (x, y) in enumerate(zip(x_coords, y_coords)):
    plt.text(x, y, f'  ({x}, {y})', verticalalignment='bottom', horizontalalignment='left', fontsize=9)

# 4. Make the grid size perfectly symmetric (same number sizing on + and - sides)
# We find the largest absolute number entered to scale the grid evenly
if len(x_coords) > 0:
    max_val = max(max(abs(x) for x in x_coords), max(abs(y) for y in y_coords), 5) + 2
else:
    max_val = 5  # Default grid range if no points are given

plt.xlim(-max_val, max_val + 1)
plt.ylim(-max_val, max_val + 1)

# 5. Draw bold center axes that cross at (0,0) to cleanly show all 4 quadrants
plt.axhline(0, color='black', linewidth=1.2)
plt.axvline(0, color='black', linewidth=1.2)

# Set equal proportions so the squares don't look stretched or distorted
plt.gca().set_aspect('equal', adjustable='box')

# 6. Customize labels, backgrounds, and layout
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Symmetric 4-Quadrant Grid')
plt.grid(True, color='lightgray', linestyle='--', linewidth=0.5)
plt.legend()

# 7. Save graph to your system
plt.savefig('symmetric_quadrant_graph.png', bbox_inches='tight')
print("\nGraph successfully saved as 'symmetric_quadrant_graph.png'!")
```
